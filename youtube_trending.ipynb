```python
import pandas as pd
from scipy import spatial
cols = ["trending_date","title","category_id","tags","views","likes","comment_count"]
# cols = ["likes"]
df = pd.read_csv("GBvideos.csv",encoding="utf-8",usecols=cols)
```

```python
df_music = df.query("category_id==10")
df_mv = df.query("category_id==26")
df_music_list = list(map(int,(df_music['likes'].tolist())))
df_mv_list = list(map(int,(df_mv['likes'].tolist())))
# print(df_music_list)
# print(df_mv_list)
print((1 - spatial.distance.cosine(df_music_list,df_mv_list)))
```
